<h4 style="color:#6c0e0eff">1. Importing NumPy, Matplotlib, Pandas, and Display Utilities from IPython</h4>

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from IPython.display import display, HTML 
```

<h4 style="color:#6c0e0eff">2. Simulating 1000 Tosses of a Fair Coin and Plotting the Relative Frequency of Heads</h4>

```python
n = 1000
twoside = np.random.choice([0, 1], size=n)
cummulative = np.cumsum(twoside)
vector = np.arange(1, n+1)
rel_freq = cummulative / vector
plt.figure(figsize=(20,5))
plt.plot(vector, rel_freq, label="Estimated P(Head)")
plt.xlabel("Number of Tosses")
plt.ylabel("Relative Frequency")
plt.title("Relative Frequency of Heads in Coin Tosses",y=1.05)
plt.axhline(0.5, linestyle="--", color='red', label="True p = 0.5")
plt.legend()
plt.grid(True)
plt.show()
```

<h4 style="color:#6c0e0eff">3. Estimating Probabilities of Heads and Tails for Different Numbers of Coin Tosses</h4>

```python
results = []
for N in [100, 1000, 10_000]:
    flips = np.random.randint(0, 2, size=N) 
    p_head = flips.mean()
    p_tail = 1 - p_head
    results.append([N, p_head, p_tail])
    
df = pd.DataFrame(results, columns=["Number of flips (N)", "Estimated P(Head)", "Estimated P(Tails)"])

html = f"""
<div style="display:flex; justify-content:center; gap:30px; margin-top:20px;">
  <div style="border:2px solid #e37e59f0; padding:15px; border-radius:8px; display:inline-block; box-shadow:2px 2px 8px #ccc;">
    <h3 style="color:#e37e59f0; text-align:center; margin-top:0;">Coin Flip Probabilities</h3>
    {df.to_html(index=False)}
</div>
"""

display(HTML(html))
```

<h4 style="color:#6c0e0eff">4. Simulating 1000 Rolls of a Fair Die and Estimating Probability of Getting a Number Greater than 3 (X>3) </h4>

```python
n = 1000  
rolls = np.random.randint(1, 7, size=n)         
event = (rolls > 3).astype(int)               
cummulative = np.cumsum(event)
trials = np.arange(1, n+1)
p = cummulative / trials
plt.figure(figsize=(20,5))
plt.plot(trials, p, label="Estimated P(X>3)")
plt.axhline(0.5, linestyle="--", color='red', label="True p = 3/6 = 0.5")
plt.xlabel("Number of rolls")
plt.ylabel("Estimated probability")
plt.title("Rolling a fair die: P(number > 3)",y=1.05)
plt.legend()
plt.grid(True)
plt.show()
```

<h4 style="color:#6c0e0eff">5. Estimating Probabilities of Outcomes Greater than 3 and Less than or Equal to 3 in Die Rolls</h4>

```python
results = []
for N in [100, 1000, 10_000]:
    rolls = np.random.randint(1, 7, size=N)
    p_gt3 = (rolls > 3).mean()
    p_le3 = 1 - p_gt3
    results.append([N, p_gt3, p_le3])

df = pd.DataFrame(results, columns=["Number of rolls (N)", "Estimated P(X>3)", "Estimated P(X3)"])

html = f"""
<div style="display:flex; justify-content:center; gap:30px; margin-top:20px;">
  <div style="border:2px solid #e37e59f0; padding:15px; border-radius:8px; display:inline-block; box-shadow:2px 2px 8px #ccc;">
    <h3 style="color:#e37e59f0; text-align:center; margin-top:0;">Die Roll Probabilities</h3>
    {df.to_html(index=False)}
</div>
"""

display(HTML(html))
```

<h4 style="color:#6c0e0eff">6. Simulating 1000 Rolls of Two Dice and Estimating Probability that the Sum Equals 5</h4>

```python
import numpy as np
import matplotlib.pyplot as plt

n = 1000
die1 = np.random.randint(1, 7, size=n)
die2 = np.random.randint(1, 7, size=n)
sums = die1 + die2

event = (sums == 5).astype(int)

cumulative = np.cumsum(event)
trials = np.arange(1, n+1)

p = cumulative / trials

plt.figure(figsize=(20,5))
plt.plot(trials, p, label="Estimated P(Sum=5)")
plt.axhline(4/36, linestyle="--", color='red', label="True p = 4/36  0.111")
plt.xlabel("Number of rolls")
plt.ylabel("Estimated probability")
plt.title("Rolling two dice: P(Sum = 5)", y=1.05)
plt.legend()
plt.grid(True)
plt.show()
```

<h4 style="color:#6c0e0eff">7. Estimating Probabilities of the Sum Being Equal to 5 or Not in Rolls of Two Dice</h4>

```python
import numpy as np
import pandas as pd
from IPython.display import display, HTML

results = []
for N in [100, 1000, 10_000]:
    die1 = np.random.randint(1, 7, size=N)
    die2 = np.random.randint(1, 7, size=N)
    sums = die1 + die2
    p_sum5 = (sums == 5).mean()
    p_not5 = 1 - p_sum5
    results.append([N, p_sum5, p_not5])

df = pd.DataFrame(results, columns=["Number of rolls (N)", "Estimated P(Sum=5)", "Estimated P(Sum5)"])

html = f"""
<div style="display:flex; justify-content:center; gap:30px; margin-top:20px;">
  <div style="border:2px solid #e37e59f0; padding:15px; border-radius:8px; display:inline-block; box-shadow:2px 2px 8px #ccc;">
    <h3 style="color:#e37e59f0; text-align:center; margin-top:0;">Dice Roll Probabilities</h3>
    {df.to_html(index=False)}
  </div>
"""

display(HTML(html))
```

<h4 style="color:#6c0e0eff">8. Simulating 1000 Tosses of a Biased Coin (P(Head)=0.4, P(Tail)=0.6) and Plotting Histogram</h4>

```python
n = 1000

flips = np.random.choice([0,1], size=n, p=[0.6, 0.4])

count_heads = np.sum(flips == 1)
count_tails = np.sum(flips == 0)

print(f"Heads count: {count_heads}  Estimated P(Head) = {count_heads/n:.3f}")
print(f"Tails count: {count_tails}  Estimated P(Tail) = {count_tails/n:.3f}")

plt.hist(flips, bins=[-0.5,0.5,1.5], rwidth=0.1, color="skyblue", edgecolor="red")
plt.xticks([0,1], ["Tail (0)", "Head (1)"])
plt.xlabel("Outcome")
plt.ylabel("Frequency")
plt.title("Histogram of 1000 Biased Coin Tosses",y=1.05)
plt.show()
```
