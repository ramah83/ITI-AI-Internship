3. **Experiment:** Rolling two fair dice, **Event:** Sum of two dice equal 5.

```python
#type your code here:
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Number of rolls
n = 1000

# Simulate rolling two dice n times
die1 = np.random.randint(1, 7, size=n)
die2 = np.random.randint(1, 7, size=n)

# Calculate the sum of each pair
sums = die1 + die2

# Mark 1 where sum == 5, else 0
sum_is_5 = np.where(sums == 5, 1, 0)

# Cumulative count of successful outcomes
cumulative_successes = np.cumsum(sum_is_5)

# Trial numbers
trials = np.arange(1, n + 1)

# Estimated probability at each trial
estimated_prob = cumulative_successes / trials

# Plotting the convergence
plt.figure(figsize=(10, 5))
plt.plot(trials, estimated_prob)
plt.axhline(y=4/36, color='red', linestyle='--', label="True P(Sum = 5) = 4/36  0.111")
plt.xlabel("Number of Trials")
plt.ylabel("Estimated Probability (Sum = 5)")
plt.title("Convergence of Estimated Probability: Sum of Two Dice = 5")
plt.legend()
plt.grid(True)
plt.show()
```
