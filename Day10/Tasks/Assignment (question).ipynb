```python
def gmm_plot(data_set , n0_components , rand_state ):
    gm = GaussianMixture(n_components=n0_components, random_state=rand_state)
    gm.fit(data_set[["x", "y"]])
    
    pred = gm.predict(data_set[["x", "y"]])
    plt.scatter(data_set["x"] , data_set["y"] , c = pred)
    plt.title("GaussianMixture")
        
    
```

```python
gmm_plot(dataset1 , 2 , 0 )
```

```python
def data_plot(dataset):
    plt.scatter(dataset["x"] , dataset["y"])
    plt.title("original")
```

```python
def cluster(data_set , no_cluster , rand_state):
    plt.figure(figsize=(50,15))
    plt.subplot(1,2,1)
    data_plot(data_set)
    plt.subplot(1,2,2)
    gmm_plot(data_set , no_cluster , rand_state )
    plt.show()
    
    
```

```python
cluster(dataset1 , 2 , 0)
```
