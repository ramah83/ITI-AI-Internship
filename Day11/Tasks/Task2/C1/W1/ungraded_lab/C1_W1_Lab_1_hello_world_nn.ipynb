<h4 style="color:#b03c12ff">1. Importing TensorFlow and NumPy Libraries</h4>

```python
import tensorflow as tf
import numpy as np
```

<h4 style="color:#b03c12ff">2. Defining Training Data Arrays for Input (xs) and Output (ys)</h4>

```python
xs = np.array([-1.0,  0.0, 1.0, 2.0, 3.0, 4.0], dtype=float)
ys = np.array([-3.0, -1.0, 1.0, 3.0, 5.0, 7.0], dtype=float)
```

<h4 style="color:#b03c12ff">3. Building and Compiling a Single-Layer Sequential Model</h4>

```python
model = tf.keras.Sequential([tf.keras.Input(shape=(1,)),tf.keras.layers.Dense(units=1)])
model.compile(optimizer='sgd', loss='mean_squared_error')
```

<h4 style="color:#b03c12ff">4. Training the Model on Sample Data and Predicting for Input 10.0</h4>

```python
model.fit(xs, ys, epochs=500)
print(f"model predicted: {model.predict(np.array([10.0]), verbose=0).item():.5f}")
```
